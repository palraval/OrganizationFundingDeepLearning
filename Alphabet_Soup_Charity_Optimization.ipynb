```python
# Imports dependencies

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint

#  Imports and reads the charity_data.csv

import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drops the non-beneficial ID columns, 'EIN' and 'NAME'.

application_df = application_df.drop(columns = ['EIN'])

application_df.head()
```

```python
# Looks at APPLICATION_TYPE value counts to identify and replaces with "Other"

application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Creates a list of application types to be replaced

application_types_to_replace = ['T9','T13','T12','T2','T25','T14','T29','T15','T17']

# Replaces in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Checks to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Looks at CLASSIFICATION values counts greater than 1

class_count = application_df['CLASSIFICATION'].value_counts()

filtered_counts = class_count[class_count > 1]

filtered_counts
```

```python
# Creates list of classifications to be replaced based on cutoff point

classifications_to_replace = class_count[class_count < 1883].index

# Replaces in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Checks to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
name_count = application_df['NAME'].value_counts()

# Creates list of names to be replaced based on cutoff point

names_to_replace = name_count[name_count <= 6].index

# Replaces in dataframe
for name in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(name,"Other")

# Checks to make sure replacement was successful
application_df['NAME'].value_counts()

```

```python
# Converts categorical data to numeric with `pd.get_dummies`

categories = pd.get_dummies(application_df)

categories = categories.astype(int)

categories.head()
```

```python
# Splits our preprocessed data into our features and target arrays

features_array = categories.drop(columns = ['IS_SUCCESSFUL'])

target_array = categories['IS_SUCCESSFUL']

# Splits the preprocessed data into a training and testing dataset

X_train, X_test, y_train, y_test = train_test_split(features_array, target_array, random_state = 24)

```

```python
# Creates a StandardScaler instance

scaler = StandardScaler()

# Fits the StandardScaler

X_scaler = scaler.fit(X_train)

# Scales the data

X_train_scaled = X_scaler.transform(X_train)

X_test_scaled = X_scaler.transform(X_test)
```

```python
# First Optimization Model


input_features = len(X_train_scaled[0])

hidden_nodes_layer_1 = 100

hidden_nodes_layer_2 = 80

hidden_nodes_layer_3 = 60

nn1 = tf.keras.models.Sequential()

# First hidden layer

nn1.add(tf.keras.layers.Dense(units = hidden_nodes_layer_1, input_dim = input_features, activation = 'relu'))

# Second hidden layer

nn1.add(tf.keras.layers.Dense(units = hidden_nodes_layer_2, activation = 'sigmoid'))

# Third hidden layer

nn1.add(tf.keras.layers.Dense(units = hidden_nodes_layer_3, activation = 'sigmoid'))

# Output layer

nn1.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Checks the structure of the model
nn1.summary()
```

```python
# Compiles the model

nn1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Trains the model

fitted_model_1 = nn1.fit(X_train_scaled, y_train, epochs = 125)
```

```python
# Evaluates the first model using the test data

model_loss, model_accuracy = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Exports our model to HDF5 file

nn1.save('Alphabet_Soup_Charity_Optimization_Model_1.h5')
```
